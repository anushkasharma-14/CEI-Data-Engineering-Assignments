Step3: Handling missing values

```python
df.isnull().sum()
```

```python
#Filling missing discount values with 0, as it indicates no discount
df["discount"] = df["discount"].fillna(0)

#Filling missing values in other columns with appropriate messages
df["what_customers_said"] = df["what_customers_said"].fillna("Not Available")
df["seller_name"] = df["seller_name"].fillna("Unknown Seller")
df["seller_information"] = df["seller_information"].fillna("Not Available")
df["videos"] = df["videos"].fillna("No Video Available")
df["variations"] = df["variations"].fillna("No Variations Available")
```

```python
df.isnull().sum()
```

```python
#Observed that even after filling missing values, whole variations column actually empty.
df["variations"].value_counts()
```

```python
#As variations column is actually empty, replacing all values with "No Variations"
df["variations"] = "No Variations"
```

```python
df["variations"].value_counts().head()
```

```python
#Observed that best_offer also have many empty dictionary values.
df["best_offer"].value_counts().head()
```

```python
#As best_offer column have many empty dictionary values, replacing all values with "No Offer"
df["best_offer"] = df["best_offer"].replace("{}", "No Offer")
```

```python
df["best_offer"].value_counts().head()
```

Step4: Basic operations (filter rows, select columns)

```python
#Filtering products with rating greater than 4
high_rated = df[df["rating"] > 4]
high_rated[["title", "rating"]].head()
```

```python
#Filtering products with discount greater than 50%
high_discount = df[df["discount"] > 50]
high_discount[["title", "discount"]].head()
```

```python
#Showing only selected columns: title, rating, initial_price, final_price
selected_columns = df[["title", "rating", "initial_price", "final_price"]]
selected_columns.head()
```

Step5: Checking Duplicate values

```python
#Checking for duplicate entries in the dataset
df.duplicated().sum()
#There are no duplicate entries in the dataset.
```

Step6: Creating new usefull columns

```python
#Creating a new column "amount_saved" to calculate the amount saved by customers based on initial price and discount percentage
df["amount_saved"] = (df["initial_price"] * df["discount"]) / 100
```

```python
#Creating a new column "top_product" to identify products which are highly rated and have good number of ratings.
df["top_product"] = (
    (df["ratings_count"] > 100) &
    (df["rating"] > 3.5)
)
```
